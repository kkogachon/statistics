### 4-2-6. 解説
- 事象の概念は意外と難しい。サイコロを１回降ったとき、標本空間は、 $\omega = \{1, 2, 3, 4, 5, 6\}$である。 
一方で事象とは取りうる状態と言える。例えば、サイコロの目が奇数であるでも良いし、サイコロの目が偶数であるとしてもよい。
サイコロを１回降ったときに考えられる事象は、$  \{1\}, \{2\}, \{3\}, \{4\}, \{5\}, \{6\},  \phi$　だけではない。  
サイコロの目が奇数である事象は、$ \{1,3, 5\} $であるし、サイコロの目が偶数である事象は、$ \{2,4,6\} $である  
- 根元事象（Elementary Event)：　単一の標本点からなりこれ以上分割できない（原子のようなもの） 
- 複合事象（Compound Event) ：　複数の標本点からなり二つ以上の根元事象に分解可能なもの

## 4-5. 条件付き確率

### 4-5-1. 条件付き確率の例

- バナナ、キウイ、いちご（どのブランドでも）が当たる確率を求めなさい。
-　トチオトメが当たる確率を求めなさい。
-　美人姫が当たる確率を求めなさい。　　

- いちごが当たったという事が分かっているものとしよう。その条件のもとで美人姫が当たる確率を求め、ベン図を書きなさい（必須）。

### 解説
<span style="font-size:125%">
$ P(A \mid B) = \dfrac{P(A\cap B)}{P(B)} $　  は変形できて  
$ P(A\cap B) = P(A\mid B) \cdot P(B)=P(B\mid A) \cdot P(A) $ とできる。
</span>

これを**乗法定理**という。
この二つの式はベイズの定理の根幹を支えると言える式である。


## 4-6. 独立性
上記4-5-2の例では、
- $ P(A\cap B) =  P(A \mid B) \cdot P(B) = P(A) \cdot P(B) $
となり、Aの事象、Bの事象は互いに影響を及ぼさないことがわかった。このことを**独立**という。

## 4-7.  ベイズの定理

### 医師の診断はベイズである
世の中の医者は、すべてベイズの定理を駆使しているといってよいだろう。
医者は患者から起こってしまった結果である「症状」を問診する。
例えば、頭痛、関節の痛み、急な高熱、首の痛みなどである。
その症状から原因を例えば「発熱の他に首の痛みがありリンパに腫れがあるからインフルエンザではなく亜急性甲状腺炎と予測する」
と診断を下す（もちろん検査もする）。
つまり、結果から原因を推定し診断しているのだ。一方で、通常の医学書は症例から病名を推定しているのではなく、
まず病名がありそ症状が詳細に書いてある。従って医者（特に若手の）は、症状を聞いたら必死に
勉強した病名の症状を思い出して、患者が罹患した病気を推測するのだ。  

今、結果（症状）Aを発熱とすると、発熱を引き起こす病気の原因Hは無数に存在する。
ただし、医学全書があるとしてすべての病気とその症状、そしてその症状が生じる確率を与えていたとする。
さらに、医学全書がその病気がそもそも罹患する確率（例、日本人の７０歳男性はおよそ１０％が癌になるなど）を与えていたとする。
すると、結果（症状）Aから最も可能性のある病名を確率的に言い当てる事が可能になる。

<span style="font-size:140%">
$$
\begin{align}
P(H_i | A) &= \dfrac{P(H_i \cap A)}{P(A)} \\ &= \dfrac{P(A|H_i)\cdot P(H_i)}{P(A)} \\ &= \dfrac{P(H_i) \cdot P(A|H_i)}{\sum P(H_i) \cdot P(A|H_i)} \tag{1}
\end{align}
$$
</span>

<span style="font-size:140%">
$ P(H_i | A) $は事後確率、 $P(H_i )$ は事前確率と呼ばれ、
$$
\dfrac{ P(A|H_i)}{\sum P(H_i) \cdot P(A|H_i)} \tag{2}
$$
</span>


は尤度と呼ばれる。  
患者から「頭痛、発熱、関節の痛み」と聞いた医者は、そもそも現代に可能性のあるあらゆる病気の可能性$P(H_i)$ を知った状況で「頭痛、発熱、関節の痛み」という症状から尤度を計算して掛けた結果、もっとも事後確率の高い（可能性のある）　「インフルエンザ」と症状を予測する。そして検査（インフルエンザキット）により診断を確定し、投薬を行うのである。

### 4-7-2. 問題
がんの診断である腫瘍マーカーの診断誤答率が５％であるとする。つまり、がん患者である人（事象$C$とする)を陽性判定（事象$A$とする)と診断できる可能性が95％、がん患者でない人（事象$C^c$)人をがん患者でないと陰性診断（事象$A^c$)できる可能性も95％ということである。
このような診断法を使って$P(C|A)$を求めよ。ただし、そもそも日本人ががんに罹患する確率を$P(C) = 0.005$とする。
結果をみると、そもそも罹患率の低い病気の診断にはただ一回の診断で下すのは危険であることが容易にわかるだろう。


```python
P_CandA=0.95 # がん患者が陽性判定
P_AcandCc=0.95 # がんでない患者が陰性判定 
PC = 0.005 # 　がん患者罹患率
P_C_A = P_CandA * PC/(P_CandA * PC + (1 - P_AcandCc )* (1 - PC) )  #条件付き確率
print(P_C_A)
```

```python
#  がん患者陽性判定率が異なる時の判定
import matplotlib.pyplot as plt

PAC = np.linspace(1,99, 100) /100

P = PC*PAC/(PC*PAC+(1-PC)*(1-PAC))

fig, ax = plt.subplots(figsize=(5,5))
ax.plot(PAC, P, 'r')
ax.set_title('陽性判定率と陽性判定精度',fontsize=18)
ax.set_xlabel('マーカーの判定精度',fontsize= 16 )
ax.set_ylabel('陽性判定率',fontsize=16)
```
